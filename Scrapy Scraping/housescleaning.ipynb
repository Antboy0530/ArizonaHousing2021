## Cleaning

```python
var['address'] = [ x.replace("\r\n","") for x in var['address']]
```

```python
var
```

```python
var['Price'] = [x.replace("\r\n","") for x in var['Price']]
```

```python
var2 = var[['address', 'Beds', 'Baths', 'Sqft', 'Price', 'Link']]
var2
```

```python
var2['Local_area'] = [ x.split(",")[1] for x in var2['address']]
var2
```

```python
var2['zipcode'] = [ x.split(",")[3].replace("AZ", "") for x in var2['address']]
var2
```

```python
var2['Price'] = [ x.replace("USD", "").replace("$", "") for x in var2['Price']]
var2
```

```python
var2.dtypes
```

```python
var2['Price'] = [ x.replace(",", "") for x in var2['Price']]
var2['Price'] = var2['Price'].apply(pd.to_numeric)
```

```python
var2['Sqft'] = [ str(x) for x in var2['Sqft']]
var2['Sqft'] = [ x.replace(",", "") for x in var2['Sqft']]
#var2['Sqft'] = var2['Sqft'].apply(pd.to_numeric)
var2.dtypes
```

```python
var2 = var2[['Price','address','Local_area', 'zipcode', 'Beds', 'Baths', 'Sqft', 'Link']]
var2.columns = ['Price','address','Local_area', 'zipcode', 'beds', 'baths', 'sqft', 'url']
```

```python
var2['address'] = [ x.split(",")[0] for x in var2['address']]
var2
```

```python
var2.to_csv('Scrapyhouses.csv',index=False)
```
