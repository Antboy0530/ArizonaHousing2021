# Cleaning Data

```python
var[var.isna().any(axis=1)]
```

```python
# keep only the rows that are not NA under sqft
var2 = var[var['sqft'].notna()]
```

```python
var2.info()
```

## clean labels

```python
var2.loc[:,'beds'] = [x.strip('bds') for x in var2.beds] #locate all rows under beds and strip 'beds'
```

```python
var2.loc[:,'baths'] = [x.strip('baths') for x in var2.baths]
```

```python
var2.loc[:,'Price'] = [x.strip('$') for x in var2.Price]
```

```python
var2
```

```python
var2['beds'] = var2['beds'].apply(pd.to_numeric)
```

```python
var2['baths'] = var2['baths'].apply(pd.to_numeric)
```

```python
var2['Price'] = var2['Price'].apply(lambda x: x.replace(",", ""))
```

```python
var2['sqft'] = var2['sqft'].apply(lambda x: x.replace(",", ""))
```
